```python
# Định hình được biểu đồ muốn vẽ --> line plot

# Chuẩn bị dữ liệu
d5 = d[(d['ProductName'] == 'Rice - Retail') & (d['Place'] == 'Istanbul')]

# tổng hợp dữ liệu theo giá trị năm, vì ở đây giá trị đang thể hiện là theo tháng
d6 = d5.groupby('Year')['Price'].mean().reset_index()

# ép dữ liệu cột Year về kiểu str
d6['Year']= d6['Year'].astype(str)
# Vẽ
plt.plot(d6['Year'], d6['Price'], color='green')
plt.show()
```
