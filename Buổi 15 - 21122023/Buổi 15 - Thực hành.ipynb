# Trực quan hoá dữ liệu bộ Food Price in Turkey

```python
# Vẽ các biểu đồ cột so sánh giá Milk (powder, infant formula) và Fuel (gas) 
# tháng 12 cuối năm năm 2019 của Ankara, Istanbul, Izmir và National Average.
```

```python
# chuẩn bị dữ liệu, lọc dữ liệu

data_1 = df[((df['ProductName'] == 'Milk (powder, infant formula) - Retail') | (df['ProductName'] == 'Fuel (gas) - Retail')) 
& (df['Year'] == 2019) & (df['Month'] == 12)]

data_1
```

```python
# Chuẩn bị dữ liệu # 3 cột: 
# Lọc dữ liệu tại tháng 12 năm 2019 ở Ankara, Istanbul, Izmir, và National Average 

data = df[(df['Month'] == 12) & (df['Year'] == 2019)].reset_index(drop=True) 
# Cột 2: Giá của sản phẩm Milk (powder, infant formula) 

y1 = data[data['ProductName'] == 'Milk (powder, infant formula) - Retail'][['Place','Price']] 
y1 = y1.reset_index(drop=True) 
y1.columns = ['Place', 'Price_Milk'] 

# Cột 3: Giá Fuel (gas) 
y2 = data[data['ProductName'] == 'Fuel (gas) - Retail'][['Place','Price']] 
y2 = y2.reset_index(drop=True) 
y2.columns = ['Place', 'Price_Fuel'] 
# Gộp thành 1 bảng 

data1 = pd.merge(y1, y2, how='inner', on='Place') 
# Vẽ 
data1.plot(x = 'Place', y = ['Price_Milk', 'Price_Fuel'], kind = 'bar')
```

Vẽ các biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail), giá Fuel (gas) 
trung bình cả nước (National Average) trong năm 2016, 2018, 2019 tại Thổ Nhĩ Kì.

```python
# Chuẩn bị dữ liệu:

# lọc dữ liệu
# palce: National Average
# năm 2016,2018, 2019
data_2= df[(df['Place'] == 'National Average') & 
((df['Year'] == 2019) |(df['Year'] == 2018)|(df['Year'] == 2016 ))].reset_index(drop=True) 



# cột giá gạo trung bình
y1 = data_2[data_2['ProductName'] == 'Rice - Retail'].mode()[]
y1 = y1.reset_index(drop=True) 
y1.columns = ['ProductName', 'Price_Fuel'] 
y1
# cột giá gas trung bình


```
