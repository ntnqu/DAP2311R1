- Sử dụng kiểm định One sample T test với hàm ttest_1samp với kiểm định hai phía
- Sử dụng kiểm định Independent T test với hàm ttest_ind với kiểm định phải
- Sử dụng kiểm định Independent T test với hàm ttest_ind với kiểm định hai phía


```python
# Nhập thư viện

import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu

df = pd.read_csv('GDPlist.csv', encoding='unicode_escape')
df.head()
```

```python
df.describe()
```

## Thực hiện các kiểm định: Trung bình GDP của các quốc gia trên thế giới là 500 tỉ usd/năm

#### GDP tính theo đơn vị triệu USD
H0: = 500 tỉ

H1: # 500

```python
stats.ttest_1samp(df['GDP (millions of US$)'], 500000)
```

pvalue=0.707 > 0.05 chưa đủ cơ sở để bác bỏ H0
-> chưa đủ cơ sở để nói 

## Thực hiện các kiểm định: GDP trung bình ở các quốc gia ở châu Âu cao hơn châu Á không


```python
# Lọc dữ liệu
# GDP trung bình ở các quốc gia ở châu Âu

df_euro = df[df['Continent'] == 'Europe']
df_euro
```

```python
# Tính GDP trung bình của Europe
df1 = df_euro.groupby('Continent')['GDP (millions of US$)'].mean().reset_index()
df1
```

```python
df_euro['GDP (millions of US$)'].mean()
```

```python
# GDP trung bình ở các quốc gia ở châu Á

df_asia = df[df['Continent'] == 'Asia']
df_asia
```

```python
stats.ttest_ind(df_euro['GDP (millions of US$)'], df_asia['GDP (millions of US$)'], equal_var=False)
```

```python
stats.ttest_ind(df_euro['GDP (millions of US$)'], df_asia['GDP (millions of US$)'], equal_var=False, alternative='greater')
```

## Thực hiện các kiểm định: GDP trung bình của các quốc gia ở châu Âu và châu Mỹ là bằng nhau

```python
df['Continent'].unique()
```

```python
# GDP trung bình ở các quốc gia ở châu Mỹ

df_SouthAmerica = df[df['Continent'] == 'South America']
df_SouthAmerica
```

```python
stats.ttest_ind(df_euro['GDP (millions of US$)'], df_SouthAmerica['GDP (millions of US$)'], equal_var=False)
```
