### Hợp nhất dữ liệu từ nhiều nguồn

```python
# tạo bảng df1, df2, df3

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3'],},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7'],},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11'],},
                   index=[8, 9, 10, 11])
```

```python
pd.concat([df1, df2, df3], axis=0)
```

```python
# hợp nhất cột, không có cột thuộc tính chung

pd.concat([df1, df2, df3], axis=1)
```

```python
left = pd.DataFrame ({'key': ['K0', 'K1', 'K2', 'K3', 'K6', 'K7'],
                      'A': ['A0', 'A1', 'A2', 'A3', 'A6', 'A7'],
                      'B': ['B0', 'B1', 'B2', 'B3', 'B6', 'B7']
                     })

left
```

```python
right = pd.DataFrame ({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                      'C': ['C0', 'C1', 'C2', 'C3','C4', 'A5'],
                      'D': ['D0', 'D1', 'D2', 'D3','D4', 'D5']
                     })

right
```

##### Cột chung (key) 
nối dựa trên cột chung có cùng tên

```python
pd.merge(left, right, how='right', on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

##### giả sử có 2 cột chung

on

```python
left = pd.DataFrame({"key1": ["K0", "K0", "K1", "K2"], 
                     "key2": ["K0", "K1", "K0", "K1"], 
                     "A": ["A0", "A1", "A2", "A3"], 
                     "B": ["B0", "B1", "B2", "B3"]}) 
left
```

```python
right = pd.DataFrame({"key1": ["K0", "K1", "K1", "K2"], 
                      "key2": ["K0", "K0", "K0", "K0"], 
                      "C": ["C0", "C1", "C2", "C3"], 
                      "D": ["D0", "D1", "D2", "D3"]}) 
right
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
left = pd.DataFrame ({'key': ['K0', 'K1', 'K2', 'K3', 'K6', 'K7'],
                      'A': ['A0', 'A1', 'A2', 'A3', 'A6', 'A7'],
                      'B': ['B0', 'B1', 'B2', 'B3', 'B6', 'B7']
                     })

left
```

```python
right = pd.DataFrame ({'key1': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                      'C': ['C0', 'C1', 'C2', 'C3','C4', 'A5'],
                      'D': ['D0', 'D1', 'D2', 'D3','D4', 'D5']
                     })

right
```

```python
a = pd.merge(left, right, how='inner', left_on='key', right_on='key1')
a
```

```python
c = pd.merge(left, right, how='right', left_on='key', right_on='key1')
c
```

```python
b = pd.merge(left, right, how='left', left_on='key', right_on='key1')
b
```
