```python
# nhiều điều kiện

# df[...]

# lọc điều kiện:
  # điều kiện 1
  # điều kiện 2
  # điều kiện 3
  # ...

# nối điều kiện:
  # &: và -> phải thoả mãn cả 2 điều kiện: điểm cao & đi học đầy đủ
  # |: hoặc -> thoả mãn 1 trong 2 điều kiện: hoặc điểm cao, hoặc đi học đầy đủ
  # (điều kiện 1) & (điều kiện 2) | (điều kiện 3)

# Lấy dữ liệu thoả mãn cả 2 điều kiện sau:
# join_year là năm 2021,
# follower_count lớn hơn hoặc bằng 10000
df[(df['join_year']==2021)&(df['follower_count']>=10000)]
```

```python
condition1 = df['join_year'] == 2021
condition2 = df['follower_count'] >= 10000
final_condition = condition1 & condition2
result_df = df[final_condition]
result_df
```

# Gom nhóm dữ liệu

```python
# gọi dữ liệu: df
# gọi hàm: .groupby (nhóm theo): .groupby('join_year')
# gọi giá trị mình muốn tính: [rating_bad]
# gọi agg fuction:
  # .mean(), .max()
  # .agg(['mean', 'max'])
```

```python
# Tìm rating_bad lớn nhất theo từng năm
df.groupby('join_year')['rating_bad'].max()
```

```python
# Tìm rating_bad trung bình và lớn nhất theo từng năm
df.groupby('join_year')['rating_bad'].agg(['mean','max'])
```

```python
# Tìm rating_bad trung bình và rating_good lớn nhất theo từng năm
df.groupby('join_year')[['rating_bad', 'rating_good']].agg({'rating_bad':'mean', 'rating_good':'max'})

# giá trị muốn tính: ['rating_bad', 'rating_good']
# chỉ định hàm tổng hợp: {'rating_bad':'mean', 'rating_good':'max}
```

```python
# Tính rating_good trung bình theo join_year và shop_location
df.groupby(['join_year', 'shop_location'])['rating_good'].mean()
```

```python
# gom nhóm dữ liệu

# df.groupby(nhóm)[giá trị].hàm()
# df.groupby(nhóm)[giá trị].agg(nhiều hàm)

# nhiều nhóm: [nhóm 1, nhóm 2, nhóm 3, ...]
# nhiều giá trị: [giá trị 1, giá trị 2, giá trị 3, ...]
# 1 hàm: .hàm(): .mean(), .max()
# nhiều hàm: .agg()
  # nhiều hàm cho cùng 1 giá trị: [hàm1, hàm 2, ...]
  # hàm khác nhau cho giá trị khác nhau: {'giá trị 1':'hàm1', 'giá trị 2':'hàm3', ...}
```
