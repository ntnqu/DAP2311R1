#### Vi·ªát h√≥a t√™n c√°c c·ªôt trong b·∫£ng d·ªØ li·ªáu: Country ü°™ Nuoc; Continent ü°™ Chauluc; GDP (millions of US$) ü°™ GDP (trieu $)


```python
# ƒë·ªïi t√™n c·ªôt

df = df.rename(columns= {'Country':'Nuoc', 'Continent':'Chauluc', 'GDP (millions of US$)':'GDP(trieu)'})
df.head()
```

#### Ch√®n th√™m m·ªôt c·ªôt ‚ÄúThanhpho‚Äù v√†o sau c·ªôt ‚ÄúNuoc‚Äù, gi√° tr·ªã ban ƒë·∫ßu l√† gi√° tr·ªã c·ªßa c·ªôt ‚ÄúNuoc‚Äù    


```python
# D√πng ph∆∞∆°ng th·ª©c insert v·ªõi tham s·ªë th·ª© 3 l√† gi√° tr·ªã c·∫ßn th√™m ƒë∆∞·ª£c x√°c ƒë·ªãnh b·∫±ng ph∆∞∆°ng th·ª©c loc: df.loc[:,‚ÄôNuoc‚Äô] 

df.insert(1, 'Thanhpho', pd.Series(df.loc[:, 'Nuoc'], index=df.index))
df.head()
```

#### Trong c·ªôt Thanhpho, thay gi√° tr·ªã Vietnam th√†nh Hanoi; L√†m t∆∞∆°ng t·ª± v·ªõi c√°c n∆∞·ªõc c√≤n l·∫°i.


```python
df['Thanhpho'].replace('Vietnam', 'Hanoi', inplace=True)
df.tail()
```

```python
df['Thanhpho'] = df['Thanhpho'].replace({'Vietnam': 'Hanoi'})
df.tail()
```
