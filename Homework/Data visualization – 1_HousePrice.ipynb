```python
# Nhập thư viện
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu

d = pd.read_csv('house_price_Dống-Da_Hà-Nội_subdata.csv', encoding='unicode_escape')
d.head()
```

```python
# Xem tổng quát thống kê mô tả
d.describe()
```

### Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, giữa số phòng ngủ với giá nhà, giữa số toilet với giá nhà.

### diện tích với giá nhà

```python
# xác định biểu đổ --> scatter plot

# chuẩn bị dữ liệu: cột diện tích và cột giá nhà

# vẽ
plt.scatter(d['area'], d['price'], color='black')
plt.title('Mối quan hệ giữa diện tích với giá nhà')
plt.xlabel('Diện tích')
plt.ylabel('Giá (tỷ đồng)')
plt.show()
```

#### Nhận xét:

Dưới diện tích 2000 thì giá cả có biến động nhiều hơn

### giữa số phòng ngủ với giá nhà

```python
plt.scatter(d['bedroom'], d['price'], color='green')
plt.title('Mối quan hệ giữa phòng ngủ với giá nhà')
plt.xlabel('Phòng ngủ')
plt.ylabel('Giá (tỷ đồng)')
plt.show()
```

#### Nhận xét:

Số lượng phòng ngủ dưới 5 thì giá cả có biến động nhiều hơn, đặc biệt là ở các căn có 1 đến 3 phòng ngủ

### giữa số toilet với giá nhà.

```python
plt.scatter(d['toilet'], d['price'], color='red')
plt.title('Mối quan hệ giữa số toilet với giá nhà')
plt.xlabel('Số toilet')
plt.ylabel('Giá (tỷ đồng)')
plt.show()
```

#### Nhận xét:

Đa số só phhongf toilet dưới 8 có sự biến động về giá nhiều hơn.
Số lượng toilet 3, 6 8 thì giá cả có biến động lớn.

###  Vẽ biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land).

```python
# Vẽ biểu đồ cột giữa 2 hình thức nhà
```

```python
# Xóa bỏ hết tất cả những dòng dữ liệu không có thông tin về giá.

d.dropna(subset = ['price'], inplace=True)
```

```python
# tạo cột 'giá/m2'
d['giá/m2'] = round((d['price']* 1000) / d['area'], 2)
d
```

```python
# gom nhóm dữ liệu theo hình thức nhà

d1 = d.groupby('property_type')['giá/m2'].mean().reset_index()
d1
```

### Vẽ biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ.

```python
# Xác định biểu đồ: cột
# cột x: số lượng phòng ngủ
# cột y: là giá trung bình .m2

# chuẩn bị dữ liệu
d2 = d.groupby('bedroom')['giá/m2'].mean().reset_index()
d2
```

```python
# vẽ

plt.bar(d2['bedroom'], d2['giá/m2'], color='green')
plt.title('Biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ')
plt.xlabel('Số phòng ngủ')
plt.ylabel('Giá (tỷ đồng/m2)')
plt.show()
```

#### Nhận xét:

Các căn có số phòng ngủ là 10 có giá trung bình /m2 cao nhất.
