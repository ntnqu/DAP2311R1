### Mục tiêu
Luyện tập xây dựng bảng Pivot table.

##### Bộ dữ liệu Online Retail chứa dữ liệu về giao dịch mua sắm online, các thông tin được thống kê bao gồm:

InvoiceNo: Số hóa đơn

StockCode: Mã kho

Description: Mô tả

Quantity: Số lượng

InvoiceDate: Ngày mua

UnitPrice: Giá

CustomerID: Mã khách hàng

Country: Quốc gia


### Yêu cầu
Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
# Nhập thư viện

import pandas as pd
import numpy as np
```

```python
# Đọc dữ liệu

df = pd.read_csv('OnlineRetail.csv', encoding='unicode_escape')
df.head()
```

### Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 


```python
# bảng dữ liệu: df
# values: tính trung bình cộng số lượng các mặt hàng -> 'Quantity'
# index: mỗi Số hóa đơn... theo từng Quốc gia -> 'Country', 'InvoiceNo'
# dggfunc: mean

df1 = df.pivot_table(values='Quantity', index=['Country', 'InvoiceNo'], aggfunc='mean')
df1
```

### Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.


```python
# với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

#values: số lượng mua hàng -> 'Quantity'
# index: mỗi Khách hàng.. theo Kho -> CustomerID, StockCode
# aggfunc: min, max

df2 = df.pivot_table(values='Quantity', index=['CustomerID', 'StockCode'], aggfunc=['min','max'])
df2
```

### Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.


```python
# với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá

# values:tổng số lượng các mặt hàng -> Quantity -> aggfunc= sum
# trung bình cộng giá -> UnitPrice -> addfunc: mean

#index: mỗi Mã kho -> StockCode

df3 = df.pivot_table(values=['Quantity', 'UnitPrice'], index='StockCode', aggfunc={'Quantity':'sum', 'UnitPrice':'mean'})
df3
```

### Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.

```python
# cho biết tổng số lượng hàng bán được của mỗi ngày

# values: Quantity	-> sum
# index: InvoiceDate

df4 = df.pivot_table(values='Quantity', index='InvoiceDate', aggfunc='sum')
df4
```
