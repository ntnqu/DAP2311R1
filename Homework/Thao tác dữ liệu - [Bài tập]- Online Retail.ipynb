```python
# Nhâp thư viện
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Đọc dữ liệu

df = pd.read_csv('OnlineRetail.csv', encoding='unicode_escape')
df.head()
```

### Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;


```python
pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceDate'].info()
```

```python
df['Month'] = pd.to_datetime(df['InvoiceDate']).dt.month
df['Month']
```

```python
df.head()
```

```python
conditions = [(df['Month'] < 3), (df['Month'] < 6), (df['Month'] < 9)]

choices = [1, 2, 3]
df['Previous'] = np.select(conditions, choices, default=4)
df.sample(10)
```

```python
df['Month'].unique()
```

### Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice


```python
df['Amount'] = df['Quantity'] * df['UnitPrice']
df.head()
```

### Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

```python
conditions_2 = [(df['Country'] == 'United Kingdom') & (df['Previous'] == 4), 
                (df['Country'] == 'France') & (df['Previous'] == 4)]

choices_2 = [0.1, 0.05]
df['Discount'] = np.select(conditions_2, choices_2, default=0)
df.sample(10)
```

### Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

```python
df['Total'] = df['Amount'] - (df['Discount']*df['Amount'])
df.sample(10)
```
